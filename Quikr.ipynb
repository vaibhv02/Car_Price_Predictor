## Cleaning

### creating a backup of the data

```python
backup=car.copy()
```

- Check if the data is numeric or not and removing the non numeric data.
- Changing car "year" column to int.

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year']=car['year'].astype(int)
```

- "Price" have Ask for Price, we are removing that and changing price from obj to int

```python
car = car[car['Price']!="Ask For Price"]
```

```python
car['Price']
```

- Removing comas by empty string and change to int.

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
```

```python
car.info()
```

- we are going to replace comas by empty string

```python
car['kms_driven']=car['kms_driven'].str.get(0).str.replace(',','')
```

- removig non numeric values

```python
car=car[car['kms_driven'].str.isnumeric()]
```

- converting obj to int

```python
car['kms_driven']= car['kms_driven'].astype(int)
```

```python
car.info()
```

- removing fuel type NAN value

```python
car[car['fuel_type'].isna()]
```

```python
car=car[~car['fuel_type'].isna()]
```

- Keep first 3 words of the name

## Model

```python
X = car.drop(columns='Price')
Y = car['Price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
Y_pred
```

```python
r2_score(Y_test,Y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,Y_train)
    Y_pred = pipe.predict(X_test)
    scores.append(r2_score(Y_test,Y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,Y_train)
Y_pred = pipe.predict(X_test)
r2_score(Y_test,Y_pred)
```
